## Calculating the mean score per region

```python
# Creating an empty dictionary
mean_happiness = {}
# Creating an array for every unique value for the Region column
unique_regions = happiness2015["Region"].unique()
# Running through unique values:
#      1. Creating a region_group with all countries from the same region
#      2. Calculating the mean score per region
#      3. Adding the country and the mean value as key-value pair to
#.        the dictionary
for region in unique_regions:
    region_group = happiness2015[happiness2015["Region"] == region]
    mean = region_group["Happiness Score"].mean()
    mean_happiness[region] = mean
```

```python
mean_happiness
```

Instead of creating a dictionary and splitting the data set into groups, executing calculations and adding the results to the dictionary, using the groupby function can be an easier and faster approach to get to the same results.

## Using the agg() method

Using the agg() method in order to execute multiple arithmetic calculations at the same time.

```python
# Calculating the mean, max, and difference of min and max for every 
# region
import numpy as np

def dif(group):
    return (group.max() - group.mean())

happy_mean_max = happy_grouped.agg([np.mean, np.max])
mean_max_dif = happy_grouped.agg([dif])
print(happy_mean_max)
mean_max_dif
```

The calculation shows that the difference between the lowest and the highest happiness score from the following reasons are the highest: Middle East and Northern Africa, and Southeastern Asia. On the contrary, Australia and New Zealand as well as North America have the lowest difference. One reason for this observation is that both region only contain two countries who in addition are quite similar to their counterpart in the regions group. It would be interesting to do more research on the reason why the difference is so high in Middle East and Northern Africa and to what extent inequality of wealth can explain these results.

# Using pivot tables

```python
# Using the pivot_table method to generate a dataframe with the 
# mean value per region.
pv_happiness = happiness2015.pivot_table(values='Happiness Score',
                            index='Region', aggfunc=np.mean, margins=True)
```

```python
pv_happiness
```

```python
# Generating a dataframe with the pivot_table method and visualize
# the results using the dataframe.plot method. 
pv_happiness = happiness2015.pivot_table('Happiness Score', 'Region')
pv_happiness.plot(kind='barh', title='Mean Happiness Scores by Region',
                 xlim=(0,10), legend=False)
```

## Groupby method and pivot_table method in comparison

```python
# Using the the groupby function to generate the min, max, and mean
# for the happiness score and the family score.
happy_family_stats = happiness2015.groupby('Region')[['Happiness Score',
                        'Family']].agg([np.min,np.max,np.mean])
happy_family_stats
```

```python
# Using the pivot_table method to generate the min, max, and mean
# for the happiness score and the family score.
pv_happy_family_stats=happiness2015.pivot_table(values=['Happiness Score','Family'],
                                            index='Region', aggfunc=[np.min, np.max,
                                            np.mean], margins=True)
pv_happy_family_stats
```

Using the groupby method and the pivot_table method shows the differences in how the data is organized and presented in the object.
